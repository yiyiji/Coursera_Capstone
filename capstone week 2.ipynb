# Car accident severity prediction

## Introduction

Car accidents are a massive global issue that often result in injury, disability, death and property damage. It has financial costs to society and individuals involved. United nations have predicted that in 2020 car accident deaths and injuries will exceed HIV/AIDS as a burden of death and disability, and the fifth leading cause of death by 2030.

A number of factors contribute to the risk of collisions, including weather conditions, road conditions, light conditions, speed of operations, use of alcohol or drugs, etc. By analyzing these factors, one could predict the severity of the accidents.

The accident severity predictions could be used to alert the drivers, the competent authorities whether a day has or not high severity factors, so the drivers could avoid driving or taking alternative routes and the authorities can prevent accidents by taking necessary measures (for instance repairing roads, improving lightening, better design intersections, etc.) or prepare for the causalities by allocating resources in advance.
